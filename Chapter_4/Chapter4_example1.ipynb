# 캐글 산탄데르 고객 만족 예측

- 고객 만족 여부. 1이면 불만족, 0이면 만족

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

customer_df = pd.read_csv('../datasets/train_santander.csv', encoding = 'latin-1')
print(customer_df.shape)
customer_df.head(3)
```

```python
customer_df.info()
```

- Null값은 X

```python
print(customer_df["TARGET"].value_counts())
unsatisfied_cnt = customer_df[customer_df["TARGET"]==1]["TARGET"].count()
total_cnt = customer_df["TARGET"].count()
print("TARGET 1의 비율은 {:.2f}".format((unsatisfied_cnt / total_cnt)))
```

```python
customer_df.describe()
```

- 위에서 var3칼럼 min값이 -99999인데 NaN값이나 특정 예외값 -99999처리 했을것이다.

```python
customer_df["var3"].value_counts()
```

- 편차를 줄여주기 위해 -999999를 가장 많은 값인 2값으로 변환시킴

```python
customer_df["var3"].replace(-999999, 2, inplace = True)
customer_df.drop("ID", axis = 1, inplace = True)

X_features = customer_df.iloc[:, :-1]
y_labels = customer_df.iloc[:, -1]
print("독립 변수 데이터 shape: {}".format(X_features.shape))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size = 0.2, random_state = 0)

train_cnt = y_train.count()
test_cnt = y_test.count()

print("학습 세트 Shape: {}, 테스트 세트 Shape: {}\n".format(X_train.shape, X_test.shape))

print(' 학습 세트 레이블 값 분포 비율')
print(y_train.value_counts() / train_cnt)

print('\n 테스트 세트 레이블 값 분포 비율')
print(y_test.value_counts() / test_cnt)
```

### XGBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb_clf = XGBClassifier(n_estimators = 500, random_state = 156)

# 성능지표 roc로 하고, 조기 중단 파라미터 100으로 설정하고 학습 수행.
xgb_clf.fit(X_train, y_train, early_stopping_rounds = 100,
            eval_metric = "auc", eval_set = [(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1], average = "macro")
print("ROC AUC: {:.4f}".format(xgb_roc_score))
```

### XGBoost GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# 수행시간이 오래걸려 100으로 낮춰서 실행
xgb_clf = XGBClassifier(n_estimators = 100)

params = {
    'max_depth':[5, 7],
    'min_child_weight':[1, 3],
    'colsample_bytree':[0.5, 0.75]
}

gridcv = GridSearchCV(xgb_clf, param_grid = params, cv = 3)
gridcv.fit(X_train, y_train, early_stopping_rounds = 30, eval_metric = "auc",
           eval_set = [(X_train, y_train), (X_test, y_test)])

print("GridSearchCV 최적 파라미터: ", gridcv.best_params_)

xgb_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:, 1], average = "macro")
print("ROC AUC: {:.4f}".format(xgb_roc_score))
```

- GridSearchCV 최적 파라미터:  {'colsample_bytree': 0.5, 'max_depth': 5, 'min_child_weight': 3} ROC AUC: 0.8445로 성능이 조금 개선됨

#### 위의 최적 파라미터로 학습

```python
xgb_clf = XGBClassifier(n_estimators = 1000, random_state = 156, learning_rate = 0.02, max_depth = 5,
                        min_child_weight = 3, colsample_bytree = 0.5, reg_alpha = 0.03)

xgb_clf.fit(X_train, y_train, early_stopping_rounds = 100,
            eval_metric = "auc", eval_set = [(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1], average = "macro")
print("ROC AUC: {:.4f}".format(xgb_roc_score))
```

- ROC AUC가 84.6으로 이전 테스트보다 살짝 향상된 결과를 나타낸다.

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1, 1, figsize = (10, 8))
plot_importance(xgb_clf, ax = ax, max_num_features = 20, height = 0.4)
```

- 가장 중요한 변수는 var38, var15로 보인다.

### LightGBM

```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier(n_estimators = 500)

evals = [(X_test, y_test)]

lgbm_clf.fit(X_train, y_train, early_stopping_rounds = 100, eval_metric = "auc",
             eval_set = evals, verbose = True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:, 1], average = "macro")
print("ROC AUC: {:.4f}".format(lgbm_roc_score))
```

### LightGBM GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# 수행시간이 오래걸려 200으로 낮춰서 실행
lgbm_clf = LGBMClassifier(n_estimator = 200)

params = {
    'num_leaves':[32, 64],
    'max_depth':[128, 160],
    'min_child_weight':[60, 100],
    'subsample':[0.8, 1]
}

gridcv = GridSearchCV(xgb_clf, param_grid = params, cv = 3)
gridcv.fit(X_train, y_train, early_stopping_rounds = 30, eval_metric = "auc",
           eval_set = [(X_train, y_train), (X_test, y_test)])

print("GridSearchCV 최적 파라미터: ", gridcv.best_params_)

lgbm_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:, 1], average = "macro")
print("ROC AUC: {:.4f}".format(lgbm_roc_score))
```

- GridSearchCV 최적 파라미터:  {'max_depth': 128, 'min_child_weight': 60, 'num_leaves': 32, 'subsample': 0.8} ROC AUC: 0.8343

#### 해당 파라미터를 LightGBM에 적용

```python
lgbm_clf = LGBMClassifier(n_estimators = 1000, num_leaves = 32, subsample = 0.8, min_child_samples = 60, max_depth = 128)

evals = [(X_test, y_test)]

lgbm_clf.fit(X_train, y_train, early_stopping_rounds = 100, eval_metric = "auc", eval_set = evals,
            verbose = True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:, 1], average = "macro")
print("ROC AUC: {:.4f}".format(lgbm_roc_score))
```

- 정확도 84.18로 측정
